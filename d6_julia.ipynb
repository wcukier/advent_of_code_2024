```julia
input = zeros(140, 140)
open("d6_input") do f
    a = readlines(f)
    global input = reduce(vcat, permutedims.(collect.(a)))
end

obstacles = input .== '#'
loc = findall(input .== '^')
loc = [loc[1][1], loc[1][2]]
visited = zeros(size(input))
turn_locs = zeros(size(input))
new_obs = zeros(size(input))
attempted_obs =  zeros(size(input))
```

```julia
dir = [-1, 0]
rot = [[0 -1]
        [1 0]]
```

```julia
function is_loop(add_obs, turn_locs, loc, dir)
    # new_visited = np.copy(visited)
    new_turns = copy(turn_locs)
    while true
        # new_visited[loc[0], loc[1]] = 1
        next_loc = loc + dir
        if ~(0 < next_loc[1] <= size(input, 1))
            return false
        end
        if ~( 0 < next_loc[2] <= size(input, 2))
            return false
        end
        if add_obs[next_loc[1], next_loc[2]] > 0
            if new_turns[loc[1], loc[2]] > 2
                return true
            end
            new_turns[loc[1], loc[2]] += 1
            dir = (dir' * rot)'
        else
            loc = next_loc
        end
    end
end
```

```julia
while true
    if sum(visited) % 10 == 0
        println(sum(visited))
    end
    visited[loc[1], loc[2]] = 1
    next_loc = loc + dir
    if ~( 0 < next_loc[1] <= size(input, 1))
        break
    end
    if ~( 0 < next_loc[2] <= size(input, 2))
        break
    end
    if obstacles[next_loc[1], next_loc[2]] > 0
        turn_locs[loc[1], loc[2]] += 1
        dir = (dir' * rot)'
    else
        if attempted_obs[next_loc[1], next_loc[2]] < 1
            add_obs = copy(obstacles)
            add_obs[next_loc[1], next_loc[2]] = 1
            if is_loop(add_obs, turn_locs, loc, (dir' * rot)')
                new_obs[next_loc[1], next_loc[2]] = 1
            end
            attempted_obs[next_loc[1], next_loc[2]] = 1
        end
        loc = next_loc

    end
end
println(Int64(sum(visited)))
println(Int64(sum(new_obs)))
```
